```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
def linear_function1(x):
    # y = 0.8x - 2
    return 0.8 * x - 2


# 0.8 * [1,2,3] - 2
# [0.8*1 - 2, 0.8*2 - 2, 0.8*3 - 2]

# y = x * 2
x_values = np.linspace(-15, 20, 35)  # [1, 2, 3]
y_values1 = linear_function1(x_values)  
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green")

plt.title("y = 0.8 * x - 2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#plt.xticks([-15, -12, -5, 0, 5, 7, 12, 14])


plt.show()
```

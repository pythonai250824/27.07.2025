```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

```python
def linear_function1(x):
    # y = 0.8x - 2
    return 0.8 * x - 2


# 0.8 * [1,2,3] - 2
# [0.8*1 - 2, 0.8*2 - 2, 0.8*3 - 2]

# y = x * 2
x_values = np.linspace(-15, 20, 35)  # [1, 2, 3]
y_values1 = linear_function1(x_values)  
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green")

plt.title("y = 0.8 * x - 2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#plt.xticks([-15, -12, -5, 0, 5, 7, 12, 14])


plt.show()
```

```python
def linear_function1(x):
    # y = -x + 5
    return -1 * x + 5

def linear_function2(x):
    # y = 2x - 1
    return 2 * x - 1


# y = x * 2
x_values = np.linspace(-15, 20, 35)  # [1, 2, 3]
y_values1 = linear_function1(x_values)  
y_values2 = linear_function2(x_values)  
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")  # x-axis
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")  # y-axis
plt.plot(x_values, y_values1, color="green", label="y = -x + 5")
plt.plot(x_values, y_values2, color="red", label="y = 2x - 1")
plt.scatter(2, 3, color='blue', zorder=5)
plt.annotate(f"(2, 3)",
             (2, 3),
             textcoords="offset points",
             xytext=(0,20),
             ha='center',
             fontsize=10,
             color='blue')
plt.legend()

plt.title("y = 0.8 * x - 2")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
#plt.xticks([-15, -12, -5, 0, 5, 7, 12, 14])


plt.show()
```

```python

##### BONUS -- python solve the equation


import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, Eq, solve

# Step 1: Define the functions
def linear_function1(x):
    return -1 * x + 5

def linear_function2(x):
    return 2 * x - 1

# Step 2: Find the intersection using sympy
x_sym = symbols('x')
eq1 = -x_sym + 5
eq2 = 2*x_sym - 1
solution = solve(Eq(eq1, eq2), x_sym)
x_intersect = float(solution[0])
y_intersect = linear_function1(x_intersect)

# Step 3: Generate x values and y values
x_values = np.linspace(-15, 20, 35)
y_values1 = linear_function1(x_values)
y_values2 = linear_function2(x_values)

# Step 4: Plot the functions
plt.axhline(0, color='black', linewidth=0.8, linestyle="--")
plt.axvline(0, color='black', linewidth=0.8, linestyle="--")
plt.plot(x_values, y_values1, color="green", label="y = -x + 5")
plt.plot(x_values, y_values2, color="red", label="y = 2x - 1")

# Step 5: Plot the intersection point
plt.scatter(x_intersect, y_intersect, color='blue', zorder=5)
plt.annotate(f"({x_intersect:.1f}, {y_intersect:.1f})",
             (x_intersect, y_intersect),
             textcoords="offset points",
             xytext=(10,10),
             ha='center',
             fontsize=10,
             color='blue')

# Step 6: Final formatting
plt.legend()
plt.title("Intersection of Two Linear Functions")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim(-10, 20)
plt.ylim(-20, 20)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```
